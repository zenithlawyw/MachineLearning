# DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#fb = pd.read_csv('../data/facebook.csv', index_col="Date")
fb = pd.read_csv('../data/FB.csv', index_col="date")
```

```python
#ms = pd.read_csv('../data/microsoft.csv', index_col="Date")
ms = pd.read_csv('../data/MS.csv', index_col="date")
```

```python
print(ms.head())
```

# Show the size of a DataFrame using "Shape"

```python
print(fb.shape)
```

```python
print(ms.shape)
```

# Show summary statistics of a DataFrame

```python
print(fb.describe())
```

```python
print(ms.describe())
```

# Locate a particular row of data using "Selection by label"

```python
fb_2015 = fb.loc['2015-01-01':'2015-12-31']
```

```python
print(fb_2015.loc['2015-03-16'])
```

```python
ms_2015 = ms.loc['2015-01-01':'2015-12-31']
```

```python
print(ms_2015.loc['2015-03-16'])
```

# Locate a particular row of data using "Selection by position"

```python
print(fb)
print(fb.iloc[0,0])
```

```python
print(ms)
print(ms.iloc[-1,0])
```

# Plot the stock data using plot() method

```python
plt.figure(figsize=(10,8))
fb['close'].plot()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
# plot only the Close price of 2016 of Microsoft, 1 line 
ms.loc['2020-01-01':'2020-12-31', 'close'].plot()
plt.show()
```
