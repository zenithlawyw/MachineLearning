# Confidence Interval

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
stock = pd.read_csv('../data/stocks/facebook.csv', index_col='Date')
stock.head()
```

# Estimate the average stock return with 90% Confidence Interval

*use log return for average stock return of the stock*

```python
stock['logReturn'] = np.log(stock['Close'].shift(-1)) - np.log(stock['Close'])
```

```python
# build 90% confidence interval for log return
sample_size = stock['logReturn'].shape[0]
sample_mean = stock['logReturn'].mean()
sample_std = stock['logReturn'].std(ddof=1) / sample_size**0.5

# left and right quantile
z_left = norm.ppf(0.05)
z_right = norm.ppf(0.95)

# upper and lower bound
interval_left = sample_mean + z_left * sample_std
interval_right = sample_mean + z_right * sample_std
```

```python
print('90% confidence interval is ', (interval_left, interval_right))
```
